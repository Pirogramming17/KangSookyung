```python
import requests
```

```python
url = "https://comic.naver.com/webtoon/weekdayList?week=thu"
response = requests.get(url)

print(response)

print(response.text[:100])
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(response.text, "html.parser")

print(soup)
print(type(soup))
```

```python
raw_webtoons = soup.select("#content > div.list_area.daily_img > ul > li")

for raw_webtoon in raw_webtoons:
		print(type(raw_webtoon))
		print(raw_webtoon, end="\n\n\n")
```

```python
for raw_webtoon in raw_webtoons:
		print(raw_webtoon.select(".thumb > a > img"))
		print(raw_webtoon.select("dt > a"))
		print()
```

```python
for raw_webtoon in raw_webtoons:
		print(raw_webtoon.select_one(".thumb > a > img"))
		print(raw_webtoon.select_one("dt > a"))
		print(raw_webtoon.select_one("dl > dd "))  
		print()
```

```python
for raw_webtoon in raw_webtoons:
		print("제목: ", raw_webtoon.select_one(".thumb > a > img").get("title"))
		print("작가: ", raw_webtoon.select_one("dd > a").text)
		print("썸네일: ", raw_webtoon.select_one(".thumb > a > img").get("src"))
		print("주소: ", raw_webtoon.select_one("dt > a").get("href"))
    #print("평점: ", raw_webtoon.select_one("dl > dd").get("style"))
    
		print()
```

#content > div.list_area.daily_img > ul > li:nth-child(1) > dl > dd:nth-child(3) > div > strong
