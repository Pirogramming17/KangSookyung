```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
search = "떡볶이"

url = f"https://www.10000recipe.com/recipe/list.html?q={search}"
```

```python
res = requests.get(url)
```

```python
soup = bs(res.text, "html.parser")
```

```python
image_src_list = [img.get("src") for img in soup.select(".common_sp_list_li .common_sp_thumb a > img")]
```
